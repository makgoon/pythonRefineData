### 데이터 타입 변경

#### 데이터 타입 완전변경, 임시변경

```python
customerData.dtypes
```

```python
customerData["EMI"] = \
    customerData.EMI.astype(str) #str, int, float 사용가능
customerData.dtypes
```

```python
##기존의 데이터 냅두고 변경하는 방법
customerData["EMI_INT"] =\
    customerData.EMI.astype(int)
customerData.dtypes
```

```python
customerData.loc[customerData.EMI == "3"]
```

```python
#임시로 변경하여 조회하는 방법
customerData.loc[customerData.EMI.astype(int)>2]
```

```python
#위 같이 조회해도 data type은 변하지 않는다.
customerData.dtypes
##타입에서 int 32, 64 와같은 값의 차이는 byte 길이, 16은 4byte, 64는 8byte를 의미함.
##int 값의 범위가 4byte를 넘어간다면 변경 필요 int 32(2^31) = 	Integer (-2147483648 to 2147483647)
```

```python
##값 다중조회
emiCondition=["2","3"]

#emi가 2, 3일때
customerData.loc[customerData.EMI.isin(emiCondition)]
```

```python
#emi가 2, 3이 아닐때
#loc는 location의 약자
customerData.loc[ ~customerData.EMI.isin(emiCondition)]
```

```python
specificIndex =[1,3]

answer1 = customerData.loc[specificIndex]
#answer1 = customerData.loc[1,3] <<<<< 이건 오류남, 2개를 검색하려면 변수로 미리 담고 써야한다.
answer2 = customerData.loc[1:3] #### loc일 때는 사람이 생각하는 1~3까지 조회, lock이 빠지면 0부터 시작하여 2번째부터 3번째가지 조회
answer3 = customerData.loc[:]

print(len(answer1))
print(len(answer2))
print(len(answer3))
```
