```python
import pandas as pd
import numpy as np
```

```python
##파일로 데이터를 출력하지 않고 개별적인 데이터를 직업 입력하는 경우
#리스트 형태로 데이터 입력
#하나의 리스트는 하나의 행을 구성
productList = ["st0001", "st0002", "st0003"]
yearweekList = ["202001", "202002", "202003"]
qtyList = [10,12,15]
```

```python
##dataframe 메소드 사용 리스트를 인수로 사용할 경우 행으로 입력

eachDf = pd.DataFrame(yearweekList)
```

```python
print(eachDf)
```

```python
#위의 경우에는 컬럼명이 입력되지 않음 따라서 컬럼명을 함께 입력
productDf = pd.DataFrame(productList, columns = ["product"])
yearweekDf = pd.DataFrame(yearweekList, columns= ["yearweek"])
qtyDf = pd.DataFrame(qtyList, columns = ["qty"])
```

```python
## 각각의 행으로 입력된 정보를 하나의 테이블로 합쳐주기
##3가지 방법 존재

### 1. concat 메소드 활용
### concat 활용시에는 인수가 dataframe형태여야 함
concatDf = pd.concat([productDf, yearweekDf, qtyDf], axis=1)
concatDf
```

```python
### 2. zip 메소드 활용
###데이터프레임을 만드는 단계에서 zip을 활용, zip에 들어가는 인수는 리스트 형태, 컬럼명은
###옵션으로 추가 가능

mergedDf = pd.DataFrame(zip(productList,
                           yearweekList,
                           qtyList),
                       columns = ["product", "yearweek", "qty"])
mergedDf.head()
```

```python
### 3. 딕셔너리 유형 변수
### key 값이 컬럼명, 테이블의 정보가 리스트 형태일때 그냥 dataframe형태로 변환하면 테이블 생성

productDict = {"product": productList,
              "yearweek": yearweekList,
              "qty": qtyList}
pd.DataFrame(productDict)
```

```python
##다중조건을 활용하는 방법
## loc 함수 이후에 산술연산자를 활용하여 조건 중첩가능

customerData.loc[ (customerData.EMI == 3)&
                  (customerData.DEVICECOUNT > 5)]
```
