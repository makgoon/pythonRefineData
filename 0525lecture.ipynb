```python
import pandas as pd
```

```python
selloutData = pd.read_csv("E:/kopocloud/dataset/kopo_channel_seasonality_new.csv")
```

```python
def monthCvt(inColumn):
    #디버깅코드
#   inColumn = "201402"
    inWeek = int(str(inColumn)[4:])

    if inWeek < 5:
        outValue =  "1월"
    elif inWeek < 9:
        outValue =  "2월"
    else:
        outValue =  "다른월"
    return outValue
```

```python
selloutData.loc[(selloutData.YEARWEEK.astype(str).str[4:].astype(int)<5),"MONTH2"]  = 1
selloutData.loc[(selloutData.YEARWEEK.astype(str).str[4:].astype(int)>=5)&
                (selloutData.YEARWEEK.astype(str).str[4:].astype(int)<10),"MONTH2"] = 2
```

```python
selloutData["MONTH"] = selloutData.YEARWEEK.apply(monthCvt)
```

```python
preFix = "1207"
def monthCvt(inColumn, appe):
    #디버깅코드
#   inColumn = "201402"
    inWeek = int(str(inColumn)[4:])

    if inWeek < 5:
        outValue =  "1월"
    elif inWeek < 9:
        outValue =  "2월"
    else:
        outValue =  "다른월"
    outValue = appe + outValue
    return outValue
```

```python
selloutData.YEARWEEK.apply(monthCvt, appe=preFix)
```
